```python
import numpy as np
from tqdm import tqdm
import pandas as pd
import json
from sklearn.cluster import KMeans
```

```python
file_path = 'vacancies_no_description.jsonl'
with open(file_path, 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
len(data)
```

```python
cur_codes = set()

for elem in data:
    cur_codes.add(elem['compensation.currencyCode'])

print('Unique codes in \'compensation.currencyCode\':')
for code in cur_codes:
    print(code)
```

```python
filtered_data = [elem for elem in data if elem.get('compensation.currencyCode') == 'RUR']
```

```python
# check only RUR

cur_codes_clean = set()

for elem in filtered_data:
    cur_codes_clean.add(elem['compensation.currencyCode'])

print('Unique codes in \'compensation.currencyCode\':')
for code in cur_codes_clean:
    print(code)
```

```python
features = []
for vacancy in data:
    features.append([vacancy['area.id']]) #сделать one-hot-encoding

X = np.array(features)

n_clusters = 20
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)

print('Clusters\' centers: ')
print(kmeans.cluster_centers_)

counts = np.bincount(kmeans.labels_)
for cluster_idx, count in enumerate(counts):
    print(f'Cluster {cluster_idx}: {count} vacancies')
```
