```python
import polars as pl
vacancy = pl.read_parquet("/home/jovyan/shares/SR004.nfs2/amaksimova/exp/hh_recsys_vacancies.pq")
```

```python
import pandas as pd
desc = pd.read_json("/home/jovyan/shares/SR004.nfs2/amaksimova/exp/desc_clear.jsonl")
desc
```

```python
vacancy = vacancy.to_pandas()
vacancy['description'] = list(desc['desc'])
```

```python
names = pd.read_json('/home/jovyan/shares/SR004.nfs2/amaksimova/exp/names.jsonl')
```

```python
vacancy
```

```python
vacancy['name'] = list(names['clean_names'])
```

```python
vacancy
```

```python
df = vacancy.groupby('name').agg({'description':lambda x: list(x)})
df
```

```python
df = df.reset_index()
df
```

```python
result = df[df['description'].apply(lambda x: len(x) > 1)]
result
```

```python
desc = list(result['description'])
ddd=[]
for i in range(len(desc)):
    q = [w.split(" ")[0] for w in desc[i]]
    qq = []
    for j in range(len(desc[i])):
        if q.count(desc[i][j].split(" ")[0])<=1:
            qq.append(desc[i][j])
    if len(qq):
        ddd.append(qq)
```

```python
dd = pd.DataFrame()
query = []
pos = []
for i in ddd:
    if len(i)>1:
        query.append(i[0])
        pos.append(i[1::])
dd['query'] = query
dd['pos'] = pos
dd
```

```python
dd.to_json('/home/jovyan/shares/SR004.nfs2/amaksimova/exp/10/tune_cool_data.jsonl',index=True)
```
