```python
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
import time
import json
```

```python
data = pd.read_csv('../data/vacancies_ru.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head()
```

```python
data['clean_vacancy_id'] = data['vacancy_id'].str.replace('v_', '')
data.head()
```

```python
# все коды проф ролей и областей
    
url = f'https://api.hh.ru/professional_roles'

response = requests.get(url)

if response.status_code == 200:
    res = response.json()['categories']
```

```python
prof_area_codes = {}

for r in res:
    prof_area_codes[r['id']] = r['name']

print(prof_area_codes)


with open('prof_area_codes.json', 'w') as file:
    json.dump(prof_area_codes, file)
```

```python
prof_roles_codes = {}

for r in res:
    roles = r['roles']
    for role in roles:
        prof_roles_codes[role['id']] = role['name']

print(prof_roles_codes)


with open('prof_roles_codes.json', 'w') as file:
    json.dump(prof_roles_codes, file)
```

```python
# маппинг областей на роли
areas_to_roles = {}

for r in res:
    areas_to_roles[r['id']] = list()
    roles = r['roles']
    for role in roles:
        areas_to_roles[r['id']].append(role['id'])
    
print(areas_to_roles)


with open('areas_to_roles.json', 'w') as file:
    json.dump(areas_to_roles, file)
```

```python
# маппинг ролей на области
roles_to_areas = {}

for area, roles in areas_to_roles.items():
    for role in roles:
        if role in roles_to_areas:
            roles_to_areas[role].append(area)
        else:
            roles_to_areas[role] = [area]

print(roles_to_areas)


with open('roles_to_areas.json', 'w') as file:
    json.dump(roles_to_areas, file)
```

```python
id_used = np.empty(shape=(0,))
id_used
```

```python
# достаю проф коды для датасета

start_time = time.time()
cnt = 0

for ind in tqdm(id_list):
    if ind in id_used:
        continue
        
    cnt += 1
    id_used = np.append(id_used, np.array(ind))

    url = f'https://api.hh.ru/vacancies/{ind}'
    
    try:
        response = requests.get(url)
    except Exception:
        continue
        
    if response.status_code != 200:
        continue
    else:
        vac = response.json()
        
        if vac['professional_roles']:
            prof_code = vac['professional_roles'][0]['id']
        else:
            continue
        
        row_id = data.index[data['clean_vacancy_id'] == str(ind)]
        
        if len(row_id) > 0:
            data.loc[row_id, 'prof_codes'] = prof_code
            
    if cnt >= 500:
        break
        
end_time = time.time()
exec_time = end_time - start_time
print('Execution time:', exec_time)
```
