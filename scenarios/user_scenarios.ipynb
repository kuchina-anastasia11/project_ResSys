```python
pred=[]
x=[]
for j,i in enumerate(action_type1):
    lsat_ind = len(i) - i[::-1].index(1) - 1
    val = action_salary1[j][len(i) - i[::-1].index(1) - 1]
    if not np.isnan(val) and lsat_ind>0 and len(action_type1[j][:lsat_ind-1])>5 and len(action_type1[j][:lsat_ind-1])<50:
        qq = action_type1[j][:lsat_ind-1]
        ww = action_salary1[j][:lsat_ind-1]
        ee={1:[],2:[],3:[]}
        ey=[0,0,0]
        meean=np.nanmean(ww)
        if not np.isnan(meean):
            for i in range(len(qq)):
                if np.isnan(ww[i]):
                    ee[qq[i]].append(meean)
                else:
                    ee[qq[i]].append(ww[i])
                
            if len(ee[1])==0:
                ey[0] = meean
            else:
                ey[0] = sum(ee[1])/len(ee[1])
            if len(ee[2])==0:
                ey[1] = meean
            else:
                ey[1] = sum(ee[2])/len(ee[2])
            if len(ee[3])==0:
                ey[2] = meean
            else:
                ey[2] = sum(ee[3])/len(ee[3])
                
            x.append(ey)
            pred.append(val)
```
