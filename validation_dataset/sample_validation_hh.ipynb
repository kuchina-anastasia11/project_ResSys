```python
import pandas as pd
import numpy as np
import random
import ast
from tqdm import tqdm
```

```python
train = pd.read_csv('../data/clean_train.csv')
train.drop('Unnamed: 0', axis=1, inplace=True)
train.head()
```

```python
test = pd.read_csv('../data/clean_test.csv')
test.drop('Unnamed: 0', axis=1, inplace=True)
test.head()
```

```python
vacancy = pd.read_csv('vacancies_full_keyskills.csv')
vacancy.head()
```

```python
def remove_duplicates(lst):
    return list(set(lst))
```

```python
def remove_empty_lists(lst):
    return len(lst) > 0
```

```python
def prepare_dataset(df):
    
    df['clear_vacancy_id_update'] = df['clear_vacancy_id'].apply(lambda x: ast.literal_eval(x))\
                                                    .apply(lambda x: [id.strip(" '[]") for id in x])
    df['action_type_update'] = df['action_type'].apply(lambda x: ast.literal_eval(x))
    
    df['vacancy_id_like'] = [
        [vacancy_id for vacancy_id, action_type in zip(vacancy_ids, action_types) if action_type == 3]
        for vacancy_ids, action_types in zip(df['clear_vacancy_id_update'], df['action_type_update'])
    ]
    
    df['vacancy_id_like'] = df['vacancy_id_like'].apply(remove_duplicates)
    
    df['vacancy_id_response'] = [
        [vacancy_id for vacancy_id, action_type in zip(vacancy_ids, action_types) if action_type == 1]
        for vacancy_ids, action_types in zip(df['clear_vacancy_id_update'], df['action_type_update'])
    ]
    
    df['vacancy_id_response'] = df['vacancy_id_response'].apply(remove_duplicates)
    df = df[df['vacancy_id_response'].apply(remove_empty_lists) 
            & df['vacancy_id_like'].apply(remove_empty_lists)]
    
    return df
```

```python
train = prepare_dataset(train)
train.head()
```

```python
test = prepare_dataset(test)
test.head()
```

```python
train_user_id = list(train['user_id'].unique())
test_user_id = list(test['user_id'].unique())
```

```python
intersect_id = list(set(train_user_id) & set(test_user_id))
```

```python
valid_train = train[train['user_id'].isin(intersect_id)]
valid_train = valid_train.groupby('user_id').tail(1)
valid_train = valid_train.reset_index()
valid_train = valid_train.drop(['index'], axis=1)
valid_train = valid_train[['user_id', 'vacancy_id_like', 'vacancy_id_response']]
valid_train.head()
```

```python
valid_test = test[test['user_id'].isin(intersect_id)]
valid_test = valid_test.reset_index()
valid_test = valid_test.drop('index', axis=1)
valid_test = valid_test[['user_id', 'vacancy_id_like', 'vacancy_id_response']]
valid_test.head()
```

```python
valid_train.to_csv('valid_train.csv', index=False)
valid_test.to_csv('valid_test.csv', index=False)
```

```python
# берем сэмпл
```

```python
sample_train = valid_train[(valid_train['vacancy_id_like'].apply(len) == 1) 
                           & (valid_train['vacancy_id_response'].apply(len) == 1)].copy()
```

```python
sample_train_like = sample_train[['user_id', 'vacancy_id_like']].copy()
sample_train_response = sample_train[['user_id', 'vacancy_id_response']].copy()
```

```python
def get_id(lst):
    if lst:
        return lst[0]

sample_train_like['vacancy_id_like'] = sample_train_like['vacancy_id_like'].apply(get_id)
sample_train_response['vacancy_id_response'] = sample_train_response['vacancy_id_response'].apply(get_id)
```

```python
merged_sample_train_like = pd.merge(sample_train_like,
                                    vacancy,
                                    left_on='vacancy_id_like',
                                    right_on='vacancy_id',
                                    how='inner')
merged_sample_train_like = merged_sample_train_like[['user_id', 'vacancy_id_like',
                                                     'clean_name', 'compensation_from',
                                                     'keySkills', 'area_regionId',
                                                     'workSchedule', 'workExperience']]
merged_sample_train_like.dropna(subset=['compensation_from'], inplace=True)
```

```python
merged_sample_train_response = pd.merge(sample_train_response,
                                    vacancy,
                                    left_on='vacancy_id_response',
                                    right_on='vacancy_id',
                                    how='inner')
merged_sample_train_response = merged_sample_train_response[['user_id', 'vacancy_id_response',
                                                     'clean_name', 'compensation_from',
                                                     'keySkills', 'area_regionId',
                                                     'workSchedule', 'workExperience']]
merged_sample_train_response.dropna(subset=['compensation_from'], inplace=True)
```

```python
merged_sample_train_like.to_csv('merged_sample_train_like.csv', index=False)
merged_sample_train_response.to_csv('merged_sample_train_response.csv', index=False)
```

```python
sample_train_ids = merged_sample_train_like['user_id'].unique()
```

```python
sample_test = valid_test[valid_test['user_id'].isin(sample_train_ids)].copy()
```

```python
sample_test.rename(columns={
    'vacancy_id_like': 'vacancy_id_like_true',
    'vacancy_id_response': 'vacancy_id_response_true'
}, inplace=True)
```

```python
sample_full_validation = pd.merge(sample_train,
                     sample_test,
                     on='user_id',
                     how='inner')
sample_full_validation = sample_full_validation[['user_id', 'vacancy_id_like_for_predict',
                        'vacancy_id_like_true',
                        'vacancy_id_response_for_predict',
                        'vacancy_id_response_true']]
sample_full_validation.head()
```

```python
def merge_columns(row):
    combined = row['vacancy_id_like'] + row['vacancy_id_response']
    return list(set(combined))

train['vacancy_id_like_response'] = train.apply(merge_columns, axis=1)
```

```python
train_session = train[train['vacancy_id_like_response'].str.len() > 10].copy()
train_session.head()
```

```python
train_session.head()
```

```python
train_session['index_test'] = train_session['vacancy_id_like_response'].str[0]
train_session['index_true'] = train_session['vacancy_id_like_response'].str[1:]
```

```python
train_session.head()
```

```python
test_session_data = []

for index, row in tqdm(train_session.iterrows(), total=len(train_session), desc='Processing rows'):
    match_row = vacancy[vacancy['vacancy_id'] == row['index_test']].copy()
    match_row['user_id'] = row['user_id']
    test_session_data.append(match_row)
```

```python
test_session_df = pd.concat(test_session_data, ignore_index=True)
test_session_df = test_session_df[['user_id', 'vacancy_id', 'clean_name',
                               'keySkills', 'compensation_from', 'area_regionId',
                               'employment', 'workExperience', 'keySkills_str']]
test_session_df.dropna(subset=['compensation_from'], inplace=True)
test_session_df.head()
```

```python
test_session_df.to_csv('test_one_session.csv', index=False)
```
